```python
import pandas as pd
```

```python
ratings = pd.read_csv('D:\\ratings.csv')
```

```python
ratings.head()
```

```python
ratings1000 = ratings.groupby(ratings.userId).count()[['rating']]
ratings1000 = ratings1000[(ratings1000.rating >= 1000 )]
ratings1000.head()
```

```python
rlifetime = ratings.groupby(ratings.userId).agg({'timestamp' : ['max', 'min']})
```

```python
rlifetime['lifetime'] = rlifetime['timestamp']['max'] - rlifetime['timestamp']['min']
```

```python
rlifetime.head()
```

```python
ratings1000.join(rlifetime, how = 'left').head()
```
